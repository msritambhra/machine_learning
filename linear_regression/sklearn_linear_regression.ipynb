```python
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
```

```python
data = np.loadtxt('data.csv',delimiter  = ',')
```

```python
x = data[:,0].reshape(-1,1)
y = data[:,1].reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_, model.intercept_
```

```python
import matplotlib.pyplot as plt
```

```python
# Prediction using m and c value
plt.scatter(x,y,label = 'original data')
m = model.coef_[0]
c = model.intercept_

y_ = m*x + c

plt.plot(x,y_,'r',label = 'hypothesis linear function')
plt.legend()
plt.show()
```

```python
# Prediction using predict function

y_pred = model.predict(x_test)


plt.scatter(x_test,y_test,label = 'testing data')
plt.plot(x_test,y_pred,'r',label = 'predicted data')
plt.legend()
plt.show()
```

```python
print("Score of the model",model.score(x_test,y_test))
```
